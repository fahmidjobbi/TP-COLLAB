```python
def extrait_comp(im):
  c1=im[:,:,0]
  c2=im[:,:,1]
  c3=im[:,:,2]
  plt.figure()
  plt.subplot(221)
  plt.imshow(im)
  plt.title('clown')
  plt.subplot(222)
  plt.imshow(c1,cmap='gray')
  plt.title('R')
  plt.subplot(223)
  plt.imshow(c2,cmap='gray')
  plt.title('V')
  plt.subplot(224)
  plt.imshow(c3,cmap='gray')
  plt.title('B')
  return c1,c2,c3

r,v,b=extrait_comp(im4)
```

```python
def echant(im,pe):
  imec=im[0:im.shape[0]:pe,0:im.shape[1]:pe]
  plt.figure()
  plt.subplot(121)
  plt.imshow(im,cmap='gray')
  plt.title('Lena')
  plt.subplot(122)
  plt.imshow(imec,cmap='gray')
  plt.title('echant'+str(pe))
  return imec 

imech2=echant(im1,12)
```

```python
#LOGIQUE
a=plt.imread('/content/gdrive/MyDrive/Colab Notebooks/images2021/triangle.bmp')
a=a[:,:,0]/255
b=plt.imread('/content/gdrive/MyDrive/Colab Notebooks/images2021/cercle.bmp')
b=b[:,:,0]/255
c=plt.imread('/content/gdrive/MyDrive/Colab Notebooks/images2021/rectangle.bmp')
c=c[:,:,0]/255

plt.close()
plt.figure()
plt.subplot(231)
plt.imshow(a,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(232)
plt.imshow(b,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(233)
plt.imshow(c,cmap='gray')
plt.xticks([]), plt.yticks([])
```

```python
d=np.logical_and(np.logical_and(a,b),c)*1
e=np.logical_or(np.logical_or(a,b),c)*1
f=np.logical_not(e)*1

plt.subplot(234)
plt.imshow(d,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(235)
plt.imshow(e,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(236)
plt.imshow(f,cmap='gray')
plt.xticks([]), plt.yticks([])
```

```python
#transformation specttrales:transforme de fourier
   #transforme de fourier
imc1=plt.imread('/content/gdrive/MyDrive/Colab Notebooks/images2021/carre1.png')
imc2=plt.imread('/content/gdrive/MyDrive/Colab Notebooks/images2021/carre2.png')

f=np.fft.fft2(imc1)
sf=np.abs(f)
sfc1=np.fft.fftshift(sf)

sfc2=np.fft.fftshift(np.abs(np.fft.fft2(imc2)))

plt.figure()
plt.subplot(221),plt.imshow(imc1, cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(imc2, cmap = 'gray')
plt.xticks([]), plt.yticks([])

plt.subplot(223),plt.imshow(sfc1, cmap='jet')
plt.title('Magnitude Spectrum carre1'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(sfc2, cmap='jet')
plt.title('Magnitude Spectrum carre2'), plt.xticks([]), plt.yticks([])

fig=plt.figure()
xx1, yy1 = np.mgrid[0:sfc1.shape[0], 0:sfc1.shape[1]]
ax = fig.gca(projection='3d')
ax.plot_surface(xx1, yy1, sfc1, cmap='jet')

fig = plt.figure()
xx2, yy2 = np.mgrid[0:sfc2.shape[0], 0:sfc2.shape[1]]
ax = fig.gca(projection='3d')
ax.plot_surface(xx2, yy2, sfc2, cmap='jet')
```
