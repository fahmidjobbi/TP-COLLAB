```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import random
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import make_blobs
data, y=make_blobs(n_samples=600,n_features=2,centers=4)

plt.plot(data[:,0],data[:,1],'r*')

```

```python
sum_of_squered_distances = []
for k in range(2,11):
   model = KMeans(n_clusters = k)
   model=model.fit(data)
sum_of_squered_distances.append(model.inertia_)

plt.figure()
plt.plot(k, sum_of_squered_distances,"b*")
plt.xlabel('K')
plt.xlabel('sum_of_squered_distances')  
plt.title("Methode for optimum K")
      


```

```python
w=[]
for i in range(2,11):
   model = KMeans(n_clusters = i)
   model=model.fit(data)
   w2=model.inertia_
   w.append(w2)
plt.plot(range(2,11),w,"-r*")
```

```python
model1 = KMeans(n_clusters=4)
model1.fit(data)
#"y_pred=kmeans.predict(data)
Label=model1.labels_
centre=model1.cluster_centers_
plt.scatter(data[:,0], data[:,1],c=Label, cmap="plasma")
 
 
 
plt.scatter(model1.cluster_centers_[:,0],model1.cluster_centers_[:,1],c='g',cmap='plasma')
plt.xlabel('Feature 0')
plt.xlabel('Feature 1')
plt.show()

```
