```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import random
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import make_blobs
data, y=make_blobs(n_samples=600,n_features=2,centers=4)

plt.plot(data[:,0],data[:,1],'r*')

```

```python
sum_of_squered_distances = []
for k in range(2,11):
   model = KMeans(n_clusters = k)
   model=model.fit(data)
sum_of_squered_distances.append(model.inertia_)

plt.figure()
plt.plot(k, sum_of_squered_distances,"b*")
plt.xlabel('K')
plt.xlabel('sum_of_squered_distances')  
plt.title("Methode for optimum K")
      


```

```python
w=[]
for i in range(2,11):
   model = KMeans(n_clusters = i)
   model=model.fit(data)
   w2=model.inertia_
   w.append(w2)
plt.plot(range(2,11),w,"-r*")
```

```python
model1 = KMeans(n_clusters=4)
model1.fit(data)
#"y_pred=kmeans.predict(data)
Label=model1.labels_
centre=model1.cluster_centers_
plt.scatter(data[:,0], data[:,1],c=Label, cmap="plasma")
 
 
 
plt.scatter(model1.cluster_centers_[:,0],model1.cluster_centers_[:,1],c='g',cmap='plasma')
plt.xlabel('Feature 0')
plt.xlabel('Feature 1')
plt.show()

```

```python
import matplotlib.cm as cm
import numpy as np
from yellowbrick.cluster import SilhouetteVisualizer
data, y = make_blobs (n_samples=500,
                    n_features=2,
                    centers=4,
)
d1 =data[:,0]
d2=data[:,1]
plt.plot (d1, d2, 'r*')
range_n_clusters =[2, 3, 4, 5, 6,7,8,9,10]
for n_clusters in range_n_clusters:
    clusterer = KMeans (n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict (data)
    silhouette_avg = silhouette_score(data, cluster_labels)
  
    visualizer = SilhouetteVisualizer(clusterer)
    visualizer.fit(data)
  
    plt.show()
                                                  
```

```python
for n_clusters in range(2,11) :
  plt.subplot(1,2,1)
  model = KMeans(n_clusters)
  visualizer = SilhouetteVisualizer (model, colors='yellowbrick')
  visualizer.fit(data)
  plt.subplot(1,2,2)
    
  plt.scatter(data[:,0],data[:,1],c=model.labels_ , cmap='plasma')
  visualizer.show()
                                
    
```

```python
'''Hierarhical clustring'''
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage (data, method='ward'))
plt.axhline(y=15, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
YPred=cluster.fit_predict(data)
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=YPred, cmap='plasma')
```
