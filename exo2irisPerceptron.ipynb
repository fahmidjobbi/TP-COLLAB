```python
from sklearn import datasets
iris=datasets.load_iris()
print(iris)
```

```python
data=iris.data
label=iris.target
print(data)
print(label)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(data,label,test_size=0.33,random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras import losses
model=Sequential()
model.add(Dense(8,input_dim=4,activation="relu"))
model.add(Dense(3,activation='softmax'))
model.compile('adam',loss=losses.mean_squared_error,metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=32,epochs=70,validation_data=(X_test,y_test))
```
