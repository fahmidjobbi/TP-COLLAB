```python
#Loading the dataset
import pandas as pd
read_file = pd.read_excel("/content/cleandata.xlsx")
read_file.to_csv ("dataclean.csv", index = None,header=True)
dataclean = pd.read_csv("/content/dataclean.csv")
print("Number of datapoints:", len(dataclean))
dataclean = dataclean.replace(np.nan,"", regex=True)
dataclean.head()
```

```python
# Elbow curve to find optimal K
cost = []
K = range(1,30)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1)
    kmode.fit_predict(dataclean)
    cost.append(kmode.cost_)
    
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Building the model with 4 clusters
kmode = KModes(n_clusters=30, init = "random", n_init = 5, verbose=1)
clusters = kmode.fit_predict(dataclean)
clusters
```

```python
def get_df_name(df):
    name =[x for x in globals() if globals()[x] is df][0]
    return name
```
