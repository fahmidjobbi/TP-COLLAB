```python
import sys
import re
import json
import os
import cv2
import numpy as np

from requests import get
from tqdm import tqdm
from bs4 import BeautifulSoup as soup
from concurrent.futures import ThreadPoolExecutor

from pydotmap import DotMap

def get_pinterest_links(body):
    searched_urls = []
    html = soup(body, 'html.parser')
    #print(soup.prettify(html))
    links = html.select('#main > div > div > div > a')
    #print(links)
    print('[+] saving results ...')
    for link in links:
        link = link.get('href')
        link = re.sub(r'/url\?q=', '', link)
        #print(link[0])
        if link[0] != "/" and "pinterest" in link:
            searched_urls.append(link)

    return searched_urls


json_data_list = []
def get_source( url):
    try:
        res = get(url)
    except Exception as e:
        return
    html = soup(res.text, 'html.parser')
        # get json data from script tag having id initial-state
    json_data = html.find_all("script", attrs={"id": "__PWS_DATA__"})
    for a in json_data:
        json_data_list.append(a.string)
       
        
        
def save_image_url():
        print('[+] saving image urls ...')
        url_list = [i for i in json_data_list if i.strip()]
        if not len(url_list):
            return url_list
        url_list = []
        for js in json_data_list:
            try:
                data = DotMap(json.loads(js))
                #print("data.props.initialReduxState.pins")
                #print(data.props.initialReduxState.pins)
                urls = []
                for pin in data.props.initialReduxState.pins:
                    if isinstance(data.props.initialReduxState.pins[pin].images.get("orig"), list):
                        for i in data.props.initialReduxState.pins[pin].images.get("orig"):
                            urls.append(i.get("url"))
                            
                    else:
                        urls.append(data.props.initialReduxState.pins[pin].images.get("orig").get("url"))

                
                for url in urls:
                    url_list.append(url)
                
            except Exception as e:
                continue
        
 
        return list(set(url_list))




ListTopic=["books","cats"]
listurl=[]
listurl+=listurl
topic=[]
topic+=topic
for k in ListTopic:
    
    
   
  keyword =   k + " pinterest"
  keyword = keyword.replace("+", "%20")
  url = f'http://www.google.co.in/search?hl=en&q={keyword}'
  print('[+] starting search ...')
  res = get(url)
  #print(res.content)
  searched_urls =get_pinterest_links(res.content)


  print('[+] saving json data ...')
  for i in searched_urls:
    get_source(i)

  url_list = save_image_url()
  listurl+=url_list
  print(len(url_list))
  print(url_list)
    
  for j in url_list:
    topic.append(k)
    
  print(topic)
```
